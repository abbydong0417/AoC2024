```python
# Read aoc10.txt as a matrix of numbers
with open('aoc10.txt') as f:
    data = f.read().splitlines()
    data = [list(map(int, x)) for x in data]
print(data)
```

```python
# Function to find all 0 values in the matrix
def find_zeros(matrix):
    zeros = []
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if matrix[i][j] == 0:
                zeros.append((i, j))
    return zeros

find_zeros(data)
```

```python
# Function to find if there is any possible path starting at 0, ending at 9, and always increasing by 1
def find_path(matrix, start, end):
    path = []
    path.append(start)
    while True:
        i, j = path[-1]
        if i == end[0] and j == end[1]:
            return path
        if i + 1 < len(matrix) and matrix[i + 1][j] == matrix[i][j] + 1:
            path.append((i + 1, j))
        elif i - 1 >= 0 and matrix[i - 1][j] == matrix[i][j] + 1:
            path.append((i - 1, j))
        elif j + 1 < len(matrix[i]) and matrix[i][j + 1] == matrix[i][j] + 1:
            path.append((i, j + 1))
        elif j - 1 >= 0 and matrix[i][j - 1] == matrix[i][j] + 1:
            path.append((i, j - 1))
        else:
            return False
```

```python
# For each zeros, find all possible paths to 9 which increases by 1, return the number of paths
def find_paths(matrix):
    zeros = find_zeros(matrix)
    count = 0
    for zero in zeros:
        if find_path(matrix, zero, (9, 9)):
            count += 1
    return count
```
