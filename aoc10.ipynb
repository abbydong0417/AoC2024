```python
# Read aoc10.txt as a matrix of numbers
with open('aoc10.txt') as f:
    data = f.read().splitlines()
    data = [list(map(int, x)) for x in data]
print(data)
```

```python
from collections import deque

# Directions for moving: up, down, left, right
directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

def bfs(map_grid, start, n, m):
    # Initialize BFS queue
    queue = deque([start])
    visited = set()
    visited.add(start)
    reachable_nines = 0

    while queue:
        x, y = queue.popleft()

        # If we reach a '9', increment the count
        if map_grid[x][y] == 9:
            reachable_nines += 1

        # Explore the four possible directions
        for dx, dy in directions:
            nx, ny = x + dx, y + dy
            if 0 <= nx < n and 0 <= ny < m and (nx, ny) not in visited:
                # We can move to a cell if it is exactly 1 greater in height
                if map_grid[nx][ny] == map_grid[x][y] + 1:
                    queue.append((nx, ny))
                    visited.add((nx, ny))
    
    return reachable_nines

def total_trailhead_score(map_grid):
    n = len(map_grid)
    m = len(map_grid[0])
    total_score = 0
    
    # Scan for trailheads (cells with height 0)
    for i in range(n):
        for j in range(m):
            if map_grid[i][j] == 0:
                # Start BFS from this trailhead
                score = bfs(map_grid, (i, j), n, m)
                total_score += score
    
    return total_score

# Example usage:
map_grid = [
    [8, 9, 0, 1, 0, 1, 2, 3],
    [7, 8, 1, 2, 1, 8, 7, 4],
    [8, 7, 4, 3, 0, 9, 6, 5],
    [9, 6, 5, 4, 9, 8, 7, 4],
    [4, 5, 6, 7, 8, 9, 0, 3],
    [3, 2, 0, 1, 9, 0, 1, 2],
    [0, 1, 3, 2, 9, 8, 0, 1],
    [1, 0, 4, 5, 6, 7, 3, 2]
]

print(total_trailhead_score(map_grid))  # Output the total score of all trailheads
```

```python
total_trailhead_score(data)  # Output the total score of all trailheads in the input data
```
