```python
# Read the file and convert it to a word matrix
with open('aoc4.txt', 'r') as file:
    word_matrix = [line.split() for line in file]

# Display the word matrix
for row in word_matrix:
    print(row)
```

```python
def find_word(word_matrix, word):
    rows = len(word_matrix)
    cols = len(word_matrix[0][0])
    word_len = len(word)
    directions = [
        (0, 1), (1, 0), (1, 1), (1, -1),  # right, down, down-right, down-left
        (0, -1), (-1, 0), (-1, -1), (-1, 1)  # left, up, up-left, up-right
    ]
    
    def is_valid(x, y):
        return 0 <= x < rows and 0 <= y < cols
    
    def search_from(x, y, dx, dy):
        for i in range(word_len):
            nx, ny = x + i * dx, y + i * dy
            if not is_valid(nx, ny) or word_matrix[nx][0][ny] != word[i]:
                return False
        return True
    
    positions = []
    for x in range(rows):
        for y in range(cols):
            for dx, dy in directions:
                if search_from(x, y, dx, dy):
                    positions.append((x, y, dx, dy))
    
    return positions

# Find the word "XMAS" in the word_matrix
positions = find_word(word_matrix, "XMAS")
print("Positions of 'XMAS':", positions)
print("Number of positions:", len(positions))
```

```python
def find_x_mas(word_matrix):
    rows = len(word_matrix)
    cols = len(word_matrix[0][0])
    pattern = "MAS"
    pattern_len = len(pattern)
    positions = []

    def is_valid(x, y):
        return 0 <= x < rows and 0 <= y < cols

    def search_x_mas(x, y):
        directions = [
            (1, 1), (1, -1), (-1, 1), (-1, -1)  # down-right, down-left, up-right, up-left
        ]
        for dx1, dy1 in directions:
            for dx2, dy2 in directions:
                if dx1 == -dx2 and dy1 == -dy2:
                    continue
                if all(is_valid(x + i * dx1, y + i * dy1) and word_matrix[x + i * dx1][0][y + i * dy1] == pattern[i] for i in range(pattern_len)) and \
                   all(is_valid(x + i * dx2, y + i * dy2) and word_matrix[x + i * dx2][0][y + i * dy2] == pattern[i] for i in range(pattern_len)):
                    positions.append((x, y))
                    return

    for x in range(rows):
        for y in range(cols):
            search_x_mas(x, y)

    return positions

# Find the X-MAS patterns in the word_matrix
x_mas_positions = find_x_mas(word_matrix)
print("Positions of 'X-MAS':", x_mas_positions)
print("Number of positions:", len(x_mas_positions))
```
