```python
# Read the file and convert it to a word matrix
with open('aoc4.txt', 'r') as file:
    word_matrix = [line.split() for line in file]

# Display the word matrix
for row in word_matrix:
    print(row)
```

```python
def find_word(word_matrix, word):
    rows = len(word_matrix)
    cols = len(word_matrix[0][0])
    word_len = len(word)
    directions = [
        (0, 1), (1, 0), (1, 1), (1, -1),  # right, down, down-right, down-left
        (0, -1), (-1, 0), (-1, -1), (-1, 1)  # left, up, up-left, up-right
    ]
    
    def is_valid(x, y):
        return 0 <= x < rows and 0 <= y < cols
    
    def search_from(x, y, dx, dy):
        for i in range(word_len):
            nx, ny = x + i * dx, y + i * dy
            if not is_valid(nx, ny) or word_matrix[nx][0][ny] != word[i]:
                return False
        return True
    
    positions = []
    for x in range(rows):
        for y in range(cols):
            for dx, dy in directions:
                if search_from(x, y, dx, dy):
                    positions.append((x, y, dx, dy))
    
    return positions

# Find the word "XMAS" in the word_matrix
positions = find_word(word_matrix, "XMAS")
print("Positions of 'XMAS':", positions)
print("Number of positions:", len(positions))
```
