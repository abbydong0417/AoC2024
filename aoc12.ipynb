```python
# Read aoc12.txt as input data map
with open("aoc12.txt", "r") as file:
    data = file.read()
print(data)
```

```python
def parse_map(garden_map):
    return [list(row) for row in garden_map.strip().split('\n')]

def get_neighbors(x, y, rows, cols):
    neighbors = []
    if x > 0: neighbors.append((x-1, y))
    if x < rows - 1: neighbors.append((x+1, y))
    if y > 0: neighbors.append((x, y-1))
    if y < cols - 1: neighbors.append((x, y+1))
    return neighbors

def flood_fill(garden_map, x, y, plant_type, visited):
    stack = [(x, y)]
    area = 0
    perimeter = 0
    rows, cols = len(garden_map), len(garden_map[0])
    
    while stack:
        cx, cy = stack.pop()
        if (cx, cy) in visited:
            continue
        visited.add((cx, cy))
        area += 1
        local_perimeter = 4
        
        for nx, ny in get_neighbors(cx, cy, rows, cols):
            if garden_map[nx][ny] == plant_type:
                stack.append((nx, ny))
                local_perimeter -= 1
        
        perimeter += local_perimeter
    
    return area, perimeter

def calculate_total_price(garden_map):
    garden_map = parse_map(garden_map)
    visited = set()
    total_price = 0
    
    for x in range(len(garden_map)):
        for y in range(len(garden_map[0])):
            if (x, y) not in visited:
                plant_type = garden_map[x][y]
                area, perimeter = flood_fill(garden_map, x, y, plant_type, visited)
                total_price += area * perimeter
    
    return total_price

# Example usage:
garden_map = """
RRRRIICCFF
RRRRIICCCF
VVRRRCCFFF
VVRCCCJFFF
VVVVCJJCFE
VVIVCCJJEE
VVIIICJJEE
MIIIIIJJEE
MIIISIJEEE
MMMISSJEEE
"""

print(calculate_total_price(garden_map))  # Output: 1930
```

```python
print(calculate_total_price(data))
```
