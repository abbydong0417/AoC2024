```python
# Reading the content of aoc5-1.txt
with open('aoc5-1.txt', 'r') as file1:
    data1 = file1.read()

# Reading the content of aoc5-2.txt
with open('aoc5-2.txt', 'r') as file2:
    data2 = file2.read()

# Display the content of the files
print("Content of aoc5-1.txt:")
print(data1)
print("\nContent of aoc5-2.txt:")
print(data2)
```

```python
# Parse the rules from data1
rules = [tuple(map(int, line.split('|'))) for line in data1.split('\n')]

# Parse the data lines from data2
data_lines = [list(map(int, line.split(','))) for line in data2.split('\n')]

def check_rules(line, rules):
    for x, y in rules:
        if x in line and y in line:
            if line.index(x) > line.index(y):
                return False
    return True

# Verify lines and get the middle number
valid_lines = [line for line in data_lines if check_rules(line, rules)]
middle_numbers = [line[len(line) // 2] for line in valid_lines]

# Output the results
print(f"Number of valid lines: {len(valid_lines)}")
print(f"Middle numbers of valid lines: {middle_numbers}")
```

```python
print(f"Sum of middle numbers: {sum(middle_numbers)}")
```

```python
invalid_lines = [line for line in data_lines if not check_rules(line, rules)]
def reorder_line(line, rules):
    for x, y in rules:
        if x in line and y in line:
            if line.index(x) > line.index(y):
                line[line.index(x)], line[line.index(y)] = line[line.index(y)], line[line.index(x)]
    return line
```

```python
sum_middle_numbers = 0
for line in invalid_lines:
    fixed_line = reorder_line(line, rules)
    while not check_rules(fixed_line, rules):
        fixed_line = reorder_line(fixed_line, rules)
    middle_number = fixed_line[len(fixed_line) // 2]
    sum_middle_numbers += middle_number
print(f"Sum of middle numbers of invalid lines: {sum_middle_numbers}")
```
