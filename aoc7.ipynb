```python
# Read the file
with open('aoc7.txt', 'r') as file:
    lines = file.readlines()

# Initialize lists
data_list_1 = []
data_list_2 = []

# Process each line
for line in lines:
    left, right = line.strip().split(':')
    data_list_1.append(left)
    data_list_2.append(right.split())

# Display the results
print("Data List 1:", data_list_1)
print("Data List 2:", data_list_2)
```

```python
# Function to 
def check_operations(data_list_1, data_list_2):
    sum_dfs = 0
    for i in range(len(data_list_1)):
        target = int(data_list_1[i])
        numbers = list(map(int, data_list_2[i]))
        
        # Check all combinations of + and * operations
        def dfs(index, current_value):
            if index == len(numbers):
                return current_value == target
            return dfs(index + 1, current_value + numbers[index]) or dfs(index + 1, current_value * numbers[index])
        
        if dfs(1, numbers[0]):
            sum_dfs += int(data_list_1[i])
            # print(f"Possible to obtain {data_list_1[i]} using operations on {data_list_2[i]}")
        else:
            continue
    return sum_dfs

check_operations(data_list_1, data_list_2)
```

```python
# Function to check operations including concatenation
def check_operations_with_concat(data_list_1, data_list_2):
    sum_dfs = 0
    for i in range(len(data_list_1)):
        target = int(data_list_1[i])
        numbers = list(map(int, data_list_2[i]))
        
        # Check all combinations of +, *, and concatenation operations
        def dfs(index, current_value):
            if index == len(numbers):
                return current_value == target
            concat_value = int(str(current_value) + str(numbers[index]))
            return (dfs(index + 1, current_value + numbers[index]) or 
                    dfs(index + 1, current_value * numbers[index]) or 
                    dfs(index + 1, concat_value))
        
        if dfs(1, numbers[0]):
            sum_dfs += int(data_list_1[i])
            # print(f"Possible to obtain {data_list_1[i]} using operations on {data_list_2[i]}")
        else:
            continue
    return sum_dfs
```

```python
check_operations_with_concat(data_list_1, data_list_2)
```
